# Solución Rubén

$$F(x)=\int_{-1}^{x}\frac{3}{2}y^2\text{d}y=\left.\frac{y^3}{2}\right|_{y=-1}^{y=x}=\frac{x^3}{2}-\frac{(-1)^3}{2}$$

>$$ U = \frac{x^3}{2}+\frac{1}{2}$$
>
> $$ U - \frac{1}{2} = \frac{x^3}{2}$$
>
> $$ 2(U - \frac{1}{2}) = x^3 $$
>
> $$ x = \sqrt[3]{2(U - \frac{1}{2})}$$
>
> $$ x = \sqrt[3]{(2U - 1)}$$

# Solución Pablo

### Media de la función de densidad

$$E[x] = \int_{-\infty}^\infty xf(x)dx$$

$$
E[x] = \int_{-1}^1 x \frac{3}{2} x^2 dx = \int_{-1}^1 \frac{3}{2} x^3 dx = [\frac{3}{8} x^4] \Bigm|_{-1}^1
= [\frac{3}{8} (1)^4 - (\frac{3}{8} (-1)^4)] = [\frac{3}{8} (1) - (\frac{3}{8} (1))] = (\frac{3}{8} - \frac{3}{8}) = \boldsymbol{0}
$$

### Método transformada inversa

$$ F(x) = \int_{-\infty}^x f(x) dx $$

**Función de distribución de probabilidad acumulada en el intervalo:**

$$ 
-1 \leq x \leq 1, F(x) = \int_{-1}^x \frac{3}{2} x^2 = [\frac{1}{2} x^3] \Bigm|_{-1}^x 
= (\frac{1}{2} x^3 - (-\frac{1}{2})) = \boldsymbol{\frac{x^3 + 1}{2}}
$$

**Inversa de la función:**

$$ \frac{x^3 + 1}{2} = U $$

$$ x^3 + 1 = 2U $$

$$ x^3 = 2U - 1 $$

$$ x = (2U - 1)^{1/3} $$

La $x$ obtenida nos dará variables aleatorias distribuidas de acuerdo a la función de distribución de probabilidad dada en el enunciado del ejercicio.

**Intervalo U**

$$ -1 \leq (2U - 1)^{1/3} \leq 1$$

$$ -1 \leq (2U - 1) \leq 1$$

$$ 0 \leq 2U \leq 2$$

$$ 0 \leq U \leq 1$$

```python
# Cantidad de términos (mismo para todos los métodos)
r1 = np.random.uniform(0, 0.6, 3)
r2 = np.random.uniform(0.6, 0.9, 3)
r3 = np.random.uniform(0.9, 1, 4)
r = [r1, r2, r3] 
# Todo los términos en un mismo vector
rc = np.concatenate(r)

# Función inversa # Para encontrar variables que distribuyan como la función.
f_inv_nan = lambda u: (2 * u - 1)**(1/3) if 0 <= u <= 1 else 0 # Devuelve Nan

f_inv = lambda u: (2 * u - 1)**(1/3) if u >= 0.5 else -(- 2 * u + 1)**(1/3)

# MÉTODOS PARA APROXIMAR LA MEDIA DE LA DISTRIBUCIÓN 

# 1. Montecarlo crudo
f_inv_u = list(map(lambda u: f_inv(u), rc))
media_montecarlo = np.mean(f_inv_u)

# 2. Método estratificado
N = 10
w = [(3/N)/0.6, (3/N)/0.3, (4/N)/0.1] # Pesos

# Función evaluada en cada elemento de la lista arrays contenida en la variable r.
# Nos devuelve la misma lista de arrays ya evaluada en la función
x = [np.array([f_inv(r[i][j]) for j in range(len(r[i]))]) for i in range(len(r))]

muestras = np.concatenate(list(map(lambda xi, wi: xi / wi, x, w)))

# 3. Método números complementarios
rc_c = 1 - rc
m_comple = list(map(lambda u: f_inv(u), np.concatenate([rc_c, rc])))
```

```python
def B_estratos(N: 'Número de estratos'):
    '''
    Esta función calcula la media de la función por partes del ejercicio anterior para el número de estratos deseado.
    '''
    # Nuestras variables aleatorias serán las que se encuentran en "rc" (para usar las mismas en todos lo métodos.)
    U2 = rc[0:N]
    i = np.arange(0, N)
    U2_est = (U2 + i) / N
    f_inv = lambda u: (2 * u - 1)**(1/3) if u >= 0.5 else -(- 2 * u + 1)**(1/3)
    f_inv_u = list(map(lambda u: f_inv(u), U2_est))
    return np.mean(f_inv_u)
```

```python
B = [2, 4, 6, 10]
mediaB = [B_estratos(B[i]) for i in range(len(B))]

print('La media con Montecarlo crudo es: ', media_montecarlo)
print('La media con el método estratificado es: ', muestras.mean())
print('La media con el método de números complementarios es: ', np.mean(m_comple))
print('La media con B estratos, (2, 4, 6, 10) respectivamente es: ', [mediaB[i] for i in range(len(mediaB))])
```

Aplicamos el método de la transformada inversa a la función de densidad dada para generar variables aleatorias que se distribuyen en base a esta función de probabilidad. Después, generamos o creamos estas variables aleatorias en los rangos que nos pedía el problema, estas son necesarias para el método estratificado, estas mismas variables aleatorias se usaron en todos los métodos para así poder tener un resultado más concreto. 

Analizando los resultados, podemos ver que todos los métodos de reducción de varianza hacen su trabajo y funcionan ya que el método de Montecarlo Crudo es el que menos se acerca al valor real de la media de la distribución, mientras que, con las mismas 10 variables aleatorias, los demás métodos (que aplican la reducción de varianza) vistos en clase se acercan mucho más a este valor real de la media. En este caso, el Método de Números Complementarios el que más se acerca al valor real de dicha media, recordando que siempre se usaron las mismas 10 variables aleatorias generadas a partir de la distribución.
